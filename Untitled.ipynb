```python
upi_trans["timestamp"] = pd.to_datetime(
    upi_trans["timestamp"],
    format="%d-%m-%Y %H:%M"
)
upi_trans.info()
```

```python
category_cols = [
    "transaction type",
    "merchant_category",
    "transaction_status",
    "sender_age_group",
    "receiver_age_group",
    "sender_state",
    "sender_bank",
    "receiver_bank",
    "device_type",
    "network_type",
    "day_of_week"
]

for col in category_cols:
    upi_trans[col] = upi_trans[col].astype("category")
upi_trans.info()
```

```python
upi_trans["year"] = upi_trans["timestamp"].dt.year

upi_trans["month"] = upi_trans["timestamp"].dt.month_name()

upi_trans["quarter"] = upi_trans["timestamp"].dt.quarter

upi_trans["day_name"] = upi_trans["timestamp"].dt.day_name()

upi_trans
```

```python
upi_trans.rename(
    columns={"amount (INR)": "amount_inr"},
    inplace=True
)
```

```python
from sqlalchemy import create_engine
engine = create_engine(
    "mysql+pymysql://root:@localhost:3306/upi_transaction_db"
)
upi_trans.to_sql(
    name="upi_trans",
    con=engine,
    if_exists="replace",
    index=False
)
```
